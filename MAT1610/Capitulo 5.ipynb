```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
fun1 = lambda x: x/np.sqrt(x**2+1)
fun2 = lambda x: x**4-x

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111)
t = np.linspace(0, 1.18,100)
ax.plot(t, fun1(t))
ax.plot(t, fun2(t))
plt.show()
```

```python
#%matplotlib notebook

n = 100

fig = plt.figure(figsize=(10,7))
#ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(111,projection='3d')
a = -4*np.pi/5
b =  np.pi*4/5
y = np.linspace(a,b, n)
x = 0.7 +0.25*np.sin(y)
t = np.linspace(0, np.pi*2, n)

zn = np.outer(x, np.sin(t))
yn = np.outer(x, np.cos(t))
xn = np.zeros_like(zn)

for i in range(len(x)):
    xn[i:i+1,:] = np.full_like(xn[0,:],y[i])

#ax1.plot(y, x)
ax2.plot_surface(xn, yn, zn,alpha=0.5)
plt.show()
```

```python
def data_for_cylinder_along_x(center_x,center_y,center_z, radius,height_x):
    xl =center_x-height_x*0.5
    xr =center_x+height_x*0.5
    x = np.linspace(xl, xr, 50)
    theta = np.linspace(0, 2*np.pi, 50)
    theta_grid, x_grid=np.meshgrid(theta, x)
    z_grid = radius*np.cos(theta_grid) + center_z
    y_grid = radius*np.sin(theta_grid) + center_y
    return x_grid,y_grid,z_grid

#fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
#Xc,Yc,Zc = data_for_cylinder_along_x(0.2,0,0, 0.5,0.1)
#ax.plot_surface(Xc, Yc, Zc, alpha=0.5, color='C04')
#Xc,Yc,Zc = data_for_cylinder_along_x(0.4,0,0, 0.5,0.1)
#ax.plot_surface(Xc, Yc, Zc, alpha=0.5,color="C04")
#plt.show()
```

```python
fig = plt.figure(figsize=(12,7))
ax1 = fig.add_subplot(111,projection='3d')
ax1.plot_surface(xn, yn, zn,alpha=0.5)

fig2 = plt.figure(figsize=(12,7))

ax2 = fig2.add_subplot(111, projection='3d')

N = 7
h = (b-a)/(N)
xcil = np.linspace(a,b,N+1)

for j in range(N):
    xcentro = (xcil[j+1]+xcil[j])*0.5
    Xc,Yc,Zc = data_for_cylinder_along_x(xcentro, 0, 0, 0.7 +0.25*np.sin(xcentro),h)
    ax2.plot_surface(Xc, Yc, Zc, color='C04', alpha=0.5)

plt.show()
```
