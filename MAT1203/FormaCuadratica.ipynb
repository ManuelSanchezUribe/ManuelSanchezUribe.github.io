```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def quadraticform(x1raw,x2raw, A):
    x1 = x1raw.copy()
    x2 = x2raw.copy()
    if np.isscalar(x1):
        return np.array([x1,x2]).dot(A.dot(np.array([x1,x2])))
    else:
        x1 = x1.reshape(x1.size)
        x2 = x2.reshape(x2.size)
        Qvals = np.zeros(x1.size)
        for i in range(x1.size):
            Qvals[i] =  np.array([x1[i],x2[i]]).dot(A.dot(np.array([x1[i],x2[i]])))
        return Qvals.reshape(x1raw.shape)
```

## Caso 1: elipse:

\begin{equation}
Q(x) = x^\top A x = 2x_1^{2} - 4 x_1 x_2 + 5x_2^{2},\qquad 
A = 
\begin{bmatrix}
5 &-2 \\
-2& 5
\end{bmatrix}
\end{equation}

```python
# %matplotlib inline
# Create a meshgrid for plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# ellipse
A = np.array([[4,-3],[-3,5]], dtype=np.float64)
Z = quadraticform(X,Y, A)
Z_48 = 48*np.ones(Z.shape)
# Plot the surface plot
fig = plt.figure(figsize=(14, 12))

# Surface plot
ax1 = fig.add_subplot(221, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
# ax1.plot_surface(X, Y, Z_48, cmap='cool', alpha=0.5)

ax1.set_title('Surface Plot')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_zlabel('Z axis')

# Surface plot
ax2 = fig.add_subplot(222, projection='3d')
ax2.plot_surface(X, Y, Z, cmap='viridis')
ax2.plot_surface(X, Y, Z_48, cmap='inferno', alpha=0.75)

ax2.set_title('Surface Plot')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
ax2.set_zlabel('Z axis')

# Surface plot
ax3 = fig.add_subplot(223, projection='3d')
ax3.plot_surface(X, Y, Z, cmap='viridis')
ax3.plot_surface(X, Y, Z_48, cmap='inferno', alpha=0.75)

ax3.set_title('Surface Plot')
ax3.set_xlabel('X axis')
ax3.set_ylabel('Y axis')
ax3.set_zlabel('Z axis')
ax3.view_init(elev=0, azim=-90) #Works!
# Contour plot
ax4 = fig.add_subplot(224)
contour = ax4.contour(X, Y, Z, cmap='viridis', levels=[48])
ax4.clabel(contour, inline=True, fontsize=10)
ax4.set_title('Contour Plot')
ax4.set_xlabel('X axis')
ax4.set_ylabel('Y axis')

plt.show()
```

## Caso 2: hiperbola:

\begin{bmatrix}

\end{bmatrix}
\end{equation}
