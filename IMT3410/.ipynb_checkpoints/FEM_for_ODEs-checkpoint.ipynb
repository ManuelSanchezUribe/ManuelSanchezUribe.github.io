# Finite Element Methods for Ordinary Differential Equations 

```python
import numpy as np
import matplotlib.pyplot as plt
# Quadrature Gauss-Legendre
from numpy.polynomial.legendre import leggauss
# 
from scipy.special import binom
```

```python
def quad_Gauss_Legendre(fun, a,b, n=None, theta=None, w=None):
    if theta is None or w is None:
        try:
            theta, w = leggauss(n)
        except:
            print("insert n")
    # quadrature computation
    sum = 0
    h = 0.5*(b-a)
    m = 0.5*(a+b)
    for k in range(len(theta)):
        sum += w[k]*fun(m+theta[k]*h)
    return h*sum
```

```python
# Example 
f = lambda x: x**6
theta, w = leggauss(4)
quad_Gauss_Legendre(f,0,1,theta=theta, w=w)
```

```python
# Bernstein-Bezier Basis in 1D on the interval [0,1]
Bhat = lambda t,n,j: binom(n,j)*(1-t)**(n-j)*t**j
tt = np.linspace(0,1,100)
nn = 5
for j in range(nn+1):
    plt.plot(tt, Bhat(tt,nn,j))
```

```python
# Bernstein-Bezier Basis in 1D on the interval [a,b]
def Bernsteinbasis(t,n,j,interval=None):
    if interval is not None:
        a,b = interval
        lambda1 = -(t-b)/(b-a)
        lambda2 = (t-a)/(b-a)
    else: 
        lambda1 = (1-t)
        lambda2 = t
    return binom(n,j)*lambda1**(n-j)*lambda2**j

def derivativeBernsteinbasis(t,n,j,interval=None): 
    '''
    Compute B'(t)
    '''
    if interval is not None:
        a,b = interval
        lambda1 = -(t-b)/(b-a)
        lambda2 = (t-a)/(b-a)
        dlambda1 = -1/(b-a)
        dlambda2 = 1/(b-a)
    else: 
        lambda1 = (1-t)
        lambda2 = t
        dlambda1 = -1
        dlambda2 = 1
    if j == 0:
        return (n)*lambda1**(n-1)*dlambda1
    elif n==j:
        return (n)*lambda2**(n-1)*dlambda2
    else: 
        return binom(n,j)*((n-j)*lambda1**(n-j-1)*lambda2**j*dlambda1 + j*lambda1**(n-j)*lambda2**(j-1)*dlambda2) 

def BernsteinInterp(t, ci, interval):
    n = len(ci)-1
    return sum([ci[j]*Bernsteinbasis(t,n,j,interval= interval) for j in range(n+1)])
    
a,b = -1,1
tt = np.linspace(a,b,100)
n = 5
for j in range(n+1):
    plt.plot(tt, Bernsteinbasis(tt,n,j, [a,b]))
```

```python
def fem_odesolver_linear(Coordinates, Elements, f, u0, n):
    N = Elements.shape[0]
    
    dof = (n+1)*N
    A = np.zeros((dof,dof),dtype=np.float64)
    c = np.zeros(dof,dtype=np.float64)
    theta, w = leggauss(n)
    for i in range(N): # loop over intervals
        m=(n+1)*i
        if i == 0:# initial condition
            A[0,0] = 1.0
            c[0]   = u0
        else: # continuity condition
            A[m, m] = 1.0
            A[m, m-1] = -1.0
        
        ta = Coordinates[Elements[i,0]]
        tb = Coordinates[Elements[i,1]]
    
        for k in range(1,n+1):
            phik = lambda t: Bernsteinbasis(t,n,k,[ta,tb])
            for j in range(0,n+1):
                dphij = lambda t: derivativeBernsteinbasis(t,n,j,[ta,tb])
                funA = lambda t: dphij(t)*phik(t)
                A[m+k,m+j] = quad_Gauss_Legendre(funA,ta,tb,theta=theta, w=w)
                                                       
            func = lambda t: f(t)*phik(t)
            c[m+k] = quad_Gauss_Legendre(func,ta,tb,theta=theta, w=w)
    print(c)
                                                       
    usol = np.linalg.solve(A,c)
    return usol
```

```python
def fem_odesolver_nonlinear(Coordinates, Elements, f, u0, n):
    # We use fix point iteratior
    N = Elements.shape[0]
    
    dof = (n+1)*N
    A = np.zeros((dof,dof),dtype=np.float64)
    
    theta, w = leggauss(n)
    for i in range(N): # loop over intervals
        m=(n+1)*i
        if i == 0:# initial condition
            A[0,0] = 1.0
        else: # continuity condition
            A[m, m] = 1.0
            A[m, m-1] = -1.0
        
        ta = Coordinates[Elements[i,0]]
        tb = Coordinates[Elements[i,1]]
    
        for k in range(1,n+1):
            phik = lambda t: Bernsteinbasis(t,n,k,[ta,tb])
            for j in range(0,n+1):
                dphij = lambda t: derivativeBernsteinbasis(t,n,j,[ta,tb])
                funA = lambda t: dphij(t)*phik(t)
                A[m+k,m+j] = quad_Gauss_Legendre(funA,ta,tb,theta=theta, w=w)
    
    # Initialize fixed point iteration
    usol = np.ones(dof,dtype=np.float64)
    usol[0] = u0
    error= 1
    tol = 1e-10
    maxiter = 100
    kk = 0
    while error> tol and kk < maxiter:
        c = np.zeros(dof,dtype=np.float64)
        for i in range(N): # loop over intervals
            ta = Coordinates[Elements[i,0]]
            tb = Coordinates[Elements[i,1]]
            m=(n+1)*i
            if i == 0:# initial condition
                c[0] = u0
            ci = usol[(n+1)*i:(n+1)*(i+1)]
            ui = lambda t: BernsteinInterp(t, ci, [ta,tb])
            for k in range(1,n+1):
                phik = lambda t: Bernsteinbasis(t,n,k,[ta,tb])
                func = lambda t: f(t, ui(t))*phik(t)
                c[m+k] = quad_Gauss_Legendre(func,ta,tb,theta=theta, w=w)
        unew = np.linalg.solve(A,c)
        error = np.linalg.norm(usol-unew)
        usol[:] = unew
        kk +=1
    return usol
```

```python
def plot_u_fem(usol, n, x, I):
    N = Elements.shape[0]
    fig, ax = plt.subplots(1,1)
    for i in range(N):
        ta = Coordinates[Elements[i,0]]
        tb = Coordinates[Elements[i,1]]
        ci = usol[(n+1)*i:(n+1)*(i+1)]
        ui = lambda t: BernsteinInterp(t, ci, [ta,tb])
        tti = np.linspace(ta,tb,5)
        ax.plot(tti, ui(tti))
    return fig
```

```python
# Nonlinear example
N = 10
Coordinates = np.linspace(0,1,N+1)
Elements = np.asarray([[i,i+1] for i in range(N)])
fnonlinear = lambda t,u: -2*t*u**2
# initial condition # y' = -2ty^2, y(0)=1 ==> y= 1/(1+t^2)
u0 = 1.0
usol = fem_odesolver_nonlinear(Coordinates, Elements, fnonlinear, u0, n=1)
```

```python
plot_u_fem(usol, n, x, I)
tt = np.linspace(0,1,100)
uexact = lambda t: 1.0/(1+t**2)
```
