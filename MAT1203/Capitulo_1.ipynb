## Graficar intersecciones de planos como solucion de sistemas lineales

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the coefficients of the planes
# Plane 1: x - 2y + z = 0
# Plane 2: y - 4z = 4
# Plane 3: -3y + 13z = -9

# Coefficients of the planes
coefficients = np.array([
    [1, -2, 1, 0],
    [0, 2, -8, 8],
    [0, -3, 13, -9]
])

# Solve the equations to find the intersection point
intersection_point = np.linalg.solve(coefficients[:, :3], coefficients[:, 3])
print(intersection_point)

# Plotting the intersection point
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the planes
x, y = np.meshgrid(np.linspace(28.99, 29.01, 10), np.linspace(15.99, 16.01, 10))
for i in range(coefficients.shape[0]):
    print(i)
    z = (-coefficients[i, 0] * x - coefficients[i, 1] * y + coefficients[i, 3]) / coefficients[i, 2]
    ax.plot_surface(x, y, z, alpha=0.5)

# Plot the intersection point
ax.scatter(intersection_point[0], intersection_point[1], intersection_point[2], color='r', s=100)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
