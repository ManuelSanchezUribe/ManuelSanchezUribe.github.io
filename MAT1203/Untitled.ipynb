```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python

x1 = np.linspace(20,4,2)
x2 = np.linspace(8,24,2)
x3 = np.linspace(0,6,2)
# x1-2*x2+x3 = 0
# 2*x2-8*x3 = 8
# -4*x1+5*x2+9*x3 = -9
X,Z = np.meshgrid(x1,x3)
Y1 = 0.5*(-X-Z)
Y2 = (-4*Z + 4)
Y3 = 0.2*(4*X - 9*Z - 9)

# plot the surface
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')


ax.plot_surface(X, Y1, Z, alpha=0.5)
ax.plot_surface(X, Y2, Z, alpha=0.5)
ax.plot_surface(X, Y3, Z, alpha=0.5)

ax.plot3D(29,16,3,'o')
plt.show()
```
