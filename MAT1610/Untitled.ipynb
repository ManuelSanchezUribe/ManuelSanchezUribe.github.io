```python
import numpy as np
import scipy as sp
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
from IPython.display import display
from IPython.core.display import HTML
```

```python
def exponential(a,x):
    return (a**x-1)/x
h =  np.linspace(1,0,100000, endpoint=False)

plt.plot(h, exponential(2,h),'C00', label='2');
plt.plot(h, exponential(3,h),'C01', label='3');
plt.plot(h, exponential(2.5,h),'C02', label='2.5');
plt.plot(h, exponential(2.75,h),'C03', label='2.75');
plt.plot(h, exponential(2.8,h),'C04', label='2.8');
plt.legend()
plt.show()
```
