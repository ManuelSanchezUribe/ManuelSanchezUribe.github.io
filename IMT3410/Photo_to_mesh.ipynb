```python
from scipy.special import binom
def B_curve(P,t):
    bt = np.zeros(2)
    n = len(P)-1

    for i, p in enumerate(P):

        bt[0] += p[0]*(1.0-t)**(n-i)*t**i*binom(n,i)
        bt[1] += p[1]*(1.0-t)**(n-i)*t**i*binom(n,i)


    return bt

def circunference(theta ,xc=[0,0],r=1.0):
    circ = np.zeros(2)
    circ[0] = xc[0]+r*np.cos(theta)
    circ[1] = xc[1]+r*np.sin(theta)
    return circ
        
```
