```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def quadraticform(x1raw,x2raw, A):
    x1 = x1raw.copy()
    x2 = x2raw.copy()
    if np.isscalar(x1):
        return np.array([x1,x2]).dot(A.dot(np.array([x1,x2])))
    else:
        x1 = x1.reshape(x1.size)
        x2 = x2.reshape(x2.size)
        Qvals = np.zeros(x1.size)
        for i in range(x1.size):
            Qvals[i] =  np.array([x1[i],x2[i]]).dot(A.dot(np.array([x1[i],x2[i]])))
        return Qvals.reshape(x1raw.shape)
```

## Caso 1: elipse:

\begin{equation}
Q(x) = x^\top A x = 2x_1^{2} - 4 x_1 x_2 + 5x_2^{2},\qquad 
A = 
\begin{bmatrix}
5 &-2 \\
-2& 5
\end{bmatrix}
\end{equation}

```python
# Create a meshgrid for plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# ellipse
A = np.array([[5,-2],[-2,5]], dtype=np.float64)
Z = quadraticform(X,Y, A)
# Plot the surface plot
fig = plt.figure(figsize=(14, 6))

# Surface plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Surface Plot')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_zlabel('Z axis')

# Contour plot
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, cmap='viridis', levels=[8,18,28,38,48, 58, 68,78,88,98])
ax2.set_title('Contour Plot')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
fig.colorbar(contour, ax=ax2, orientation='vertical')

plt.show()
```

## Caso 2: hiperbola:

\begin{equation}
Q(x) = x^\top A x = 2x_1^{2} - 2 x_1 x_2 -4 x_2^{2},\qquad 
A = 
\begin{bmatrix}
5 &-1 \\
-1& -4
\end{bmatrix}
\end{equation}

```python
# Create a meshgrid for plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# ellipse
A = np.array([[5,-1],[-1,-4]], dtype=np.float64)
Z = quadraticform(X,Y, A)
# Plot the surface plot
fig = plt.figure(figsize=(14, 6))

# Surface plot
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, Z, cmap='viridis')
ax1.set_title('Surface Plot')
ax1.set_xlabel('X axis')
ax1.set_ylabel('Y axis')
ax1.set_zlabel('Z axis')

# Contour plot
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, cmap='viridis', levels=[-40,-30,-20,-10,0,10,20,30,40])
ax2.set_title('Contour Plot')
ax2.set_xlabel('X axis')
ax2.set_ylabel('Y axis')
fig.colorbar(contour, ax=ax2, orientation='vertical')

plt.show()
```
