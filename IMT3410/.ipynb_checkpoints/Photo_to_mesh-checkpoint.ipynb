```python
from scipy.special import binom
def B_curve(P,t):
    bt = np.zeros(2)
    n = len(P)-1

    for i, p in enumerate(P):

        bt[0] += p[0]*(1.0-t)**(n-i)*t**i*binom(n,i)
        bt[1] += p[1]*(1.0-t)**(n-i)*t**i*binom(n,i)


    return bt
        
```

```python
P = [np.array([0.0,0.0]), np.array([0.5,1.0]),np.array([1.0,.0])]
print(P)
t = np.linspace(0,1,100)
b= B_curve(P,0.2)
b = np.asarray([ B_curve(P,ti) for ti in t])
b
```
