```python
import numpy as np
import scipy as sp
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
from IPython.display import display
from IPython.core.display import HTML
```

# Capitulo 3: Aplicaciones de la derivada

## Ejemplo: $f(x) = \frac{2x^2}{x^2-1}$

```python
epsilon=10**(-1)
x1 = np.linspace(-4,-1-epsilon,50)
x2 = np.linspace(-1+epsilon, 1-epsilon, 100)
x3 = np.linspace(1+epsilon,4,50)
x = np.linspace(-4,4,20)
f = lambda x: 2*x**2/(x**2-1)
ah = lambda x: 2+0*x
plt.figure(figsize=(8,6))
plt.plot(x1,f(x1),'C03')
plt.plot(x2,f(x2),'C03')
plt.plot(x3,f(x3),'C03')
plt.plot(x,ah(x), '--')
plt.plot([-1,-1],[-9,12], 'C00',linestyle='--')
plt.plot([1,1],[-9,12],'C00',linestyle='--')

plt.show()
```

## Ejemplo: $f(x) = \frac{x^{2}}{\sqrt{x+1}}$

```python
epsilon=1.5*10**(-2)

x = np.linspace(-1+epsilon,4,200)
f = lambda x: x**2/np.sqrt(x+1)
plt.figure(figsize=(8,6))
plt.plot(x,f(x),'C03')
plt.plot([-1,-1],[-1,8], 'C00',linestyle='--')
plt.show()
```

## Ejemplo: $f(x) = x e^{x}$

```python
epsilon=1.5*10**(-2)

x = np.linspace(-5,1,200)
f = lambda x: x*np.exp(x)
plt.figure(figsize=(8,6))
plt.plot(x,f(x),'C03')
ah = lambda x: 0*x
plt.plot(x, ah(x), 'C00',linestyle='--')
plt.show()
```

## Ejemplo: $f(x) = \frac{\cos(x)}{2+\sin(x)}$

```python
epsilon=1.5*10**(-2)

x = np.linspace(-4*np.pi,4*np.pi,200)
f = lambda x: np.cos(x)/(2+np.sin(x))
plt.figure(figsize=(16,6))
plt.plot(x,f(x),'C03')
ah = lambda x: 0*x
plt.plot([0,0], [-1/np.sqrt(3), 1/np.sqrt(3)], 'C00',linestyle='--')
plt.plot([2*np.pi,2*np.pi], [-1/np.sqrt(3), 1/np.sqrt(3)], 'C00',linestyle='--')
plt.plot([-2*np.pi,-2*np.pi], [-1/np.sqrt(3), 1/np.sqrt(3)], 'C00',linestyle='--')
plt.plot([4*np.pi,4*np.pi], [-1/np.sqrt(3), 1/np.sqrt(3)], 'C00',linestyle='--')
plt.plot([-4*np.pi,-4*np.pi], [-1/np.sqrt(3), 1/np.sqrt(3)], 'C00',linestyle='--')
plt.show()
```

## Ejemplo: $f(x) = \ln(4-x^{2})$

```python
epsilon=10**(-2)

x = np.linspace(-2+epsilon,2-epsilon,200)
f = lambda x: np.log(4-x**2)
plt.figure(figsize=(8,6))
plt.plot(x,f(x),'C03')
ah = lambda x: 0*x
plt.plot([-2,-2], [-3.5, 1.5], 'C00',linestyle='--')
plt.plot([2,2], [-3.5, 1.5], 'C00',linestyle='--')

plt.show()
```

## Ejemplo: $f(x) = \frac{x^3}{x^2+1}$

```python

x = np.linspace(-5,5,200)
f = lambda x: x**3/(x**2+1)
plt.figure(figsize=(8,6))
plt.plot(x,f(x),'C03')
ao = lambda x: 1*x
plt.plot(x, ao(x), 'C00',linestyle='--')

plt.show()
```
